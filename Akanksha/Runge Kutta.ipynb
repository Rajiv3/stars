```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as spc
```

```python
def runge_kutta(y0, t0, h, f):
    
    '''
    t0 = initial t value
    y0 = y(t0) = initial y value
    h = 'step size'
    f = 'function'
    '''
    
    k1 = f(y0,t0)
    k2 = f(y0 + 0.5*h*k1, t0 + 0.5*h)
    k3 = f(y0 + 0.5*h*k2, t0 + 0.5*h)
    k4 = f(y0 + h*k3, t0 + h)

    return y0 + (k1 + 2*k2 + 2*k3 + k4)*h/6


'''
Note: Using an example function to understand the working of runge_kutta
'''
def f(y, t):
    return (y+t)*np.sin(y*t)
```

```python
h = 0.2
t = np.arange(0, 6, h)
y = []
y0 = 5

for i in t:
    val = runge_kutta(y0, i, h, f)
    #print(str(i)+ ' - ' + str(val))
    y.append(val)
    y0 = val
    
plt.figure(figsize=(10, 9))
plt.plot(t, y, label = 'h = 0.2')
plt.xlabel('t', fontsize=20)
plt.ylabel('y', fontsize=20)
plt.legend(fontsize = 20)
```

```python
h = 0.1
t = np.arange(0, 6, h)
y = []
y0 = 5
for i in t:
    val = runge_kutta(y0, i, h, f)
    #print(str(i)+ ' - ' + str(val))
    y.append(val)
    y0 = val
    
plt.figure(figsize=(10, 9))    
plt.plot(t, y, label = 'h = 0.1')
plt.xlabel('t', fontsize=20)
plt.ylabel('y', fontsize=20)
plt.legend(fontsize = 20)
```
